```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import networkx as nx
import pandas as pd
import numpy as np
```

```python
df_7734_tumor = pd.read_csv('./out/7734_tumor/count_network.txt',sep='\t').sort_values(by=['SOURCE','TARGET'])
```

```python
cmap = plt.cm.plasma_r
plt.clf()
fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True,figsize=(15,15))
#7734_tumor
df_7734_tumor = pd.read_csv('./out/7734_tumor/count_network.txt',sep='\t').sort_values(by=['SOURCE','TARGET'])
g_7734t = nx.from_pandas_edgelist(df_7734_tumor, 'SOURCE', 'TARGET', ['count']).to_undirected()
edges,weights = zip(*nx.get_edge_attributes(g_7734t,'count').items())
pos = nx.circular_layout(g_7734t)
node_opts = {"node_size": 1000, "node_color": ['#ff7f0e','#279e68','#d62728','#8c564b'], "linewidths": 2.0}
plt.subplot(221)
plt.title("7734-PE",fontweight='bold',fontsize=25)
plt.axis('off')
nx.draw_networkx_nodes(g_7734t, pos, **node_opts)
#nx.draw_networkx_labels(g_7734t, pos, font_size=19,clip_on=False,font_weight='bold',verticalalignment='top')
edges = nx.draw_networkx_edges(g_7734t, pos, width=20, edge_color=weights,edge_cmap=cmap,alpha=0.8,
                               arrows=True,arrowstyle='-')
#7734_ptc
df_7734_ptc = pd.read_csv('./out/7734_ptc/count_network.txt',sep='\t').sort_values(by=['SOURCE','TARGET'])
g_7734c = nx.from_pandas_edgelist(df_7734_ptc, 'SOURCE', 'TARGET', ['count']).to_undirected()
plt.subplot(222)
plt.title("7734-ptc",fontweight='bold',fontsize=25)
plt.axis('off')
edges,weights = zip(*nx.get_edge_attributes(g_7734c,'count').items())
pos = nx.circular_layout(g_7734c)
node_opts = {"node_size": 1000, "node_color": ['#ff7f0e','#279e68','#d62728','#8c564b'], "linewidths": 2.0}
nx.draw_networkx_nodes(g_7734c, pos, **node_opts)
#nx.draw_networkx_labels(g_7734c, pos, font_size=19,clip_on=False,font_weight='bold',verticalalignment='top')
edges = nx.draw_networkx_edges(g_7734c, pos, width=20, edge_color=weights,edge_cmap=cmap,alpha=0.8,
                               arrows=True,arrowstyle='-')

##### set colorbar ########
pc = mpl.collections.PatchCollection(edges, cmap=cmap)
pc.set_array(weights)
cax = fig.add_axes([1,1,0.2,0.02])     #### position
cbar = plt.colorbar(pc,cax=cax,orientation='horizontal')
cbar.set_ticks([50,250])
cbar.set_ticklabels([])
#cbar.set_ticklabels(['Low','High'])
#cbar.ax.set_title('Number of ligand-receptor',fontsize=20)
cbar.ax.tick_params(labelsize=20)
for tick in cbar.ax.xaxis.get_major_ticks():
    tick.label2.set_fontweight('bold')


#0129_tumor
df_0129_tumor = pd.read_csv('./out/0129_tumor/count_network.txt',sep='\t').sort_values(by=['SOURCE','TARGET'])
g_0129t = nx.from_pandas_edgelist(df_0129_tumor, 'SOURCE', 'TARGET', ['count']).to_undirected()
plt.subplot(223)
plt.title("0129-tumor",fontweight='bold',fontsize=25)
plt.axis('off')
edges,weights = zip(*nx.get_edge_attributes(g_0129t,'count').items())
pos = nx.circular_layout(g_0129t)
node_opts = {"node_size": 1000, "node_color": ['#ff7f0e','#279e68','#d62728','#8c564b'], "linewidths": 2.0}
nx.draw_networkx_nodes(g_0129t, pos, **node_opts)
#nx.draw_networkx_labels(g_0129t, pos, font_size=19,clip_on=False,font_weight='bold',verticalalignment='top')
edges = nx.draw_networkx_edges(g_0129t, pos, width=20, edge_color=weights,edge_cmap=cmap,alpha=0.8,
                               arrows=True,arrowstyle='-')

#0129_ptc
df_0129_ptc = pd.read_csv('./out/0129_ptc/count_network.txt',sep='\t').sort_values(by=['SOURCE','TARGET'])
g_0129c = nx.from_pandas_edgelist(df_0129_ptc, 'SOURCE', 'TARGET', ['count']).to_undirected()
plt.subplot(224)
plt.title("0129-ptc",fontweight='bold',fontsize=25)
plt.axis('off')
edges,weights = zip(*nx.get_edge_attributes(g_0129c,'count').items())
pos = nx.circular_layout(g_0129c)
node_opts = {"node_size": 1000, "node_color": ['#ff7f0e','#279e68','#d62728','#8c564b'], "linewidths": 2.0}
nx.draw_networkx_nodes(g_0129c, pos, **node_opts)
#nx.draw_networkx_labels(g_0129c, pos, font_size=19,clip_on=False,font_weight='bold',verticalalignment='top')
edges = nx.draw_networkx_edges(g_0129c, pos, width=20, edge_color=weights,edge_cmap=cmap,alpha=0.8,
                               arrows=True,arrowstyle='-')
#plt.show()

plt.savefig('./lr_compare_test.png',dpi=600,facecolor='white', transparent=False,bbox_inches = 'tight')
```
